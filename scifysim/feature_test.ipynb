```python
import numpy as np

def frq2time(frq, tmax=False):
    """
    Description:
    Converts evenly spaced frequencies into sampling times.
    
    """
    
    # Warning: code is not idempotent
    n = 2 * (frq.shape[0] - 1)
    dt = 1./ (n * np.abs(frq[1]-frq[0]))
    if tmax :
        n = 1./( np.min(frq) * dt ) - 2.
    times = np.arange(n + 2,dtype=np.float64) * dt + dt/2.
    return times

def time2frq(time, fmax=False):
    dt = np.abs(time[1]-time[0]).astype(np.float64)
    n = time.shape[0] - 2
    if (n % 2 is 0):
        n = n-1
    df = 1. / (n * dt)
    
    # Max frequency from time array
    f2 = (n - 1.)/2. + 1.
    
    # add frequencies to upper limit
    if fmax:
        if f2 > fmax :
            n1 = np.ceil( (fmax - f2) / df)
            logit.info("Padding uppe frequencies with %d channels."%(n1))
            n = n + n1
    # endif else fmax = f2            ; potentially provide as output
    # !RL No, that is too dirty, I refuse to do it.
    freqs = np.arange(n/2. + 1, dtype=np.float64) * df
    return freqs

def test_t2f():
    n = 11
    print("Sequence 1")
    freq = np.arange(n)
    print(freq)
    time = frq2time(freq)
    
```
